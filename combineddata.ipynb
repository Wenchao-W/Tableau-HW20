```python
# Add dependencies
import glob
import pandas as pd
import datetime as dt
```

```python
oneyeardata = pd.concat([pd.read_csv(f) for f in glob.glob('rawdata/oneyear/*.csv')], ignore_index = True)
quaterdata=pd.concat([pd.read_csv(f) for f in glob.glob('rawdata/quater/*.csv')], ignore_index = True)
```

```python
oneyeardata.head()
```

```python
# dt.strptime(oneyeardata['starttime'][0], '%b %d %Y %I:%M%p')
from dateutil import parser
oneyeardata['Date'] = [datetime.datetime.date(parser.parse(i)).strftime('%#m/%#d/%y') for i in oneyeardata['starttime']]
```

```python
oneyeardata['age']=[(2019-i) for i in oneyeardata['birth year']]
```
